```python
import numpy as np
import plotly.graph_objects as go
```

```python
import math
from math import sqrt, sin, cos, atan2

def NeelSkyrmion(charge, polarity, cellSize):
    w = 8 * cellSize
    w2 = w*w 

    def func(x, y, z):
        r2 = x * x + y * y 
        r = sqrt(r2)
        phi = atan2(y, x)
        mx = sin(r) * cos(charge * phi + polarity)
        my = sin(r) * sin(charge * phi + polarity)
        mz = cos(charge * phi + polarity)
        s = sqrt(mx **2 + my**2 + mz**2)
        return [mx/s, my/s, mz/s]
    return func


f = NeelSkyrmion(1, 0, 1e-9)


xs = []
ys = []
zs = []
us = []
vs = []
ws = []

n = 10

z= 0 
for r in np.linspace(1, 5, 5):
    for angle in np.linspace(0, 360, 20):
        x = r * math.cos(angle / 180 * math.pi)
        y = r * math.sin(angle / 180 * math.pi)
        mx, my, mz = f(x,y,z)
        xs.append(x)
        ys.append(y)
        zs.append(z)
        us.append(mx)
        vs.append(my)
        ws.append(mz)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Cone(x=xs, y=ys, z=zs, u=us, v=vs, w=ws))
fig.update_layout(scene_camera_eye=dict(x=-0.76, y=1.8, z=0.92))
fig.show()
```
